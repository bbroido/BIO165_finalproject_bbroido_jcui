```python
# import libraries 

# standard libraries
import os
import pandas as pd
```

```python
bulk_data = pd.read_csv("neuronal_transcriptomic.csv")
display(bulk_data)

# all samples
columns = bulk_data.columns.unique()
num_rows = bulk_data.shape[0]
print("There are " + str(len(columns)) + " samples")
print(f"There are {num_rows} genes")



```

<font color = "orange">

- Context (RecallOnly) is not fear conditioned but put through a recall context
- Fearonly is fear conditioned but not put through a recal context
- Homecage is a double negative control
- FC (FearConditioned) is a double positive control

```python
# filtering through all the different scRNA data and only keep the relevant four conditions listed above
relevant_conditions = ["Fearonly", "Homecage", "Context", "FC", "Unnamed"]

filtered_list = []

for label in columns:
    for condition in relevant_conditions:
        if condition in label:
            filtered_list.append(label)
            break

print(filtered_list)
print(len(filtered_list))

# simplify the dataframe to only include the columns in filtered_list
bulk_data_filtered = bulk_data.filter(items=filtered_list)

# rename Unnamed to gene name
bulk_data_filtered.rename(columns={"Unnamed: 0": "GeneName"}, inplace=True)

display(bulk_data_filtered)
```

<font color = "orange">

Now we transpose this dataframe such that the columns are the different genes and the rows are the different samples measured.

```python
gene_name_transposed = bulk_data_filtered.set_index("GeneName").T
gene_name_transposed = gene_name_transposed.reset_index().rename(columns={"index": "GeneName"})

display(gene_name_transposed)
```

<font color = "orange">

Now we will seperate this new data frame into four, one for each specific testing condition (Fearonly, Context, Homecage, and Fear Conditioned (FC)).

```python
# split the above data frame into four, one for each condition tested
df_fearonly = gene_name_transposed[gene_name_transposed["GeneName"].str.contains("Fearonly", na=False)]
df_context = gene_name_transposed[gene_name_transposed["GeneName"].str.contains("Context", na=False)]
df_homecage = gene_name_transposed[gene_name_transposed["GeneName"].str.contains("Homecage", na=False)]
df_fc = gene_name_transposed[gene_name_transposed["GeneName"].str.contains("FC", na=False)]

df_dict = {
    "Fearonly_Genes.csv": df_fearonly,
    "Context_Genes.csv": df_context,
    "Homecage_Genes.csv": df_homecage,
    "FC_Genes.csv": df_fc
}

# save as csv files
for filename, dataframe in df_dict.items():
    display(dataframe) 
    dataframe.to_csv(filename, index=False)


```
